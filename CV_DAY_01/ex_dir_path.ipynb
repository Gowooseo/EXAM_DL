### 폴더 및 파일 다루기


```python
### 모듈 로딩
import os
```

```python
### 특정 폴더 내의 모든 데이터 리스트 가져오기
my_folder='../data/'

#해당 폴더 또는 파일이 존재하는 지 검사
os.path.exists(my_folder)
```

```python
# 폴더 내 모든 파일,폴더 리스트 변환
datalist=os.listdir(my_folder)  #listdir은 이름만 준다.

for data_name in datalist:
    print(data_name,os.path.isdir(my_folder+data_name),sep=' --- ') #is dir에는 경로를 저장해야 한다
```

```python
#data 폴더 아래에 파일은 파일명만 출력, 폴더라면 폴더 아래 데이터 리스트 출력
datalist=os.listdir(my_folder)  #listdir은 이름만 준다.
# '.csv' 확장자 파일 찾기
for data_name in datalist:
    # 폴더 여부 체크 후 폴더 일 경우 
    if os.path.isdir(my_folder+data_name): #
        sublist = os.listdir(my_folder+data_name)
        for sub in sublist:print(f'---{sub}')
    # 폴더가 아닌 경우
    else:
        print(data_name)
        
     
   
```

```python
#data 폴더 아래에 파일은 파일명만 출력, 폴더라면 폴더 아래 데이터 리스트 출력,자기가 원하는 확장자만 출력(예:csv)

# import os
# 
my_folder='../data/'
#data 폴더 아래에 파일은 파일명만 출력, 폴더라면 폴더 아래 데이터 리스트 출력
datalist=os.listdir(my_folder)  #listdir은 이름만 준다.
# '.csv' 확장자 파일 찾기
for data_name in datalist:
    # 폴더 여부 체크 후 폴더 일 경우 
    if os.path.isdir(my_folder+data_name): #
        sublist = os.listdir(my_folder+data_name)
        for sub in sublist:
            if sub.endswith('.csv'):
                print(sub)
    
            


    
```

```python
#함수 만들기(수제로 만듦)
import os

def getfiles(dir,extension):
    #data 폴더 아래에 파일은 파일명만 출력, 폴더라면 폴더 아래 데이터 리스트 출력
    datalist=os.listdir(dir)  #listdir은 이름만 준다.
    # print(datalist)
    # '.csv' 확장자 파일 찾기
    for data_name in datalist:
        # 폴더 여부 체크 후 폴더 일 경우 
        if os.path.isdir(data_name):  #폴더인지 파일인지 체크
            getfiles(dir+data_name,extension)
            
                    
        else: #만약 파일이라면 ....
            if data_name.endswith(extension):  #data_name이 친 확장자로 끝난다면....
                print(data_name)
            
            
            #시간을 잡아먹었으니까 설계도를 그려라....
            
            
```

```python
#원하는 확장자만 가져오는 def
import os

def getFiles2(folder, extension):
    file_list = []
    for root, dirs, files in os.walk(folder):
        for file in files:
            if file.endswith(extension):
                file_path = os.path.join(root, file)
                file_list.append(file_path)
    return file_list
#함수 호출 부
my_folder = "../data/"
txt_files = getFiles2(my_folder, ".jpg")

for file_path in txt_files:
    print(file_path)
```
